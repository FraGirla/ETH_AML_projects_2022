```python
import pandas as pd
import numpy as np
from sklearn import svm
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import catboost as cb
```

```python
X = pd.read_csv('X_preprocessed.csv',index_col=['id'])
```

```python
y = pd.read_csv('y_preprocessed.csv',index_col=['id'])
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state = 100)
```

```python
rfr = ExtraTreesRegressor(n_jobs=1, max_depth=None, n_estimators=180, random_state=0, min_samples_split=3,
                          max_features=None)

rfr.fit(x_train, np.ravel(y_train))

y_pred = rfr.predict(x_test)
y_pred = np.rint(y_pred)

score = r2_score(y_test, y_pred)
print(score)
```

```python
model = cb.CatBoostRegressor()
grid = {'iterations': [50, 100, 150],
        'learning_rate': [0.03, 0.1, 0.5],
        'depth': [1, 2, 4, 6],
        'l2_leaf_reg': [0.5, 1, 3, 4, 5]}
model.grid_search(grid, x_train, np.ravel(y_train))
pred = model.predict(x_test)
pred = np.rint(pred)
rmse = (np.sqrt(mean_squared_error(y_test, pred)))
r2 = r2_score(y_test, pred)
r2
```

```python
regr = svm.SVR()
regr.fit(x_train, np.ravel(y_train))
y_pred = rfr.predict(x_test)
y_pred = np.rint(y_pred)

score = r2_score(y_test, y_pred)
print(score)
```
